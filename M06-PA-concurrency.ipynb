```python
from datetime import date, datetime

today_date = str(date.today())

with open("today.txt", "wt") as today:
    today.write(today_date)
```

```python
with open("today.txt", "rt") as read_today:
    today_string = read_today.read()
    raw_date = today_string.strip()
```

```python
parsed_date = datetime.strptime(raw_date, "%Y-%m-%d").date()

print(parsed_date)
```

```python
import multiprocessing
import time
import random
from datetime import datetime

def show_time():
    # wait a random number of seconds between 0 and 1
    delay = random.random()
    time.sleep(delay)

    # print the current time
    now = datetime.now().strftime("%H:%M:%S.%f")
    print(f"Process {multiprocessing.current_process().pid} waited {delay:.3f}s → {now}")

if __name__ == "__main__":
    processes = []

    # create 3 processes
    for i in range(3):
        p = multiprocessing.Process(target=show_time)
        processes.append(p)
        p.start()

    # wait for all to finish
    for p in processes:
        p.join()
```
