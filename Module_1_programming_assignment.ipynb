How many seconds are in an hour?

```python
hours = 1
minutes_in_hour = 60
seconds_in_minutes = 60
seconds_per_hour = hours * minutes_in_hour * seconds_in_minutes
print(seconds_per_hour)
```

There's 3600 seconds in an hour. This code simply works by creating three variables to store the amount of hours, the amount of minutes in an hour, and the amount of seconds in a minute, to then proceed to multiply all of these values and storing the result in a variable called "seconds_per_hour". Finally, we print the result

How many seconds are in a day?

```python
seconds_per_day = seconds_per_hour * 24
print(seconds_per_day)
```

There's 86400 seconds in a day, this is just creating a new variable and assigning the result of the previous variable "seconds_per_hour" and multiplying that value by the 24 hours which we have in a day. Then we print that variable.

```python
print(seconds_per_day / seconds_per_hour)
```

Then we just do float division with seconds_per_day by seconds_per_hour and print it out

```python
print(seconds_per_day // seconds_per_hour)
```

And we finish by doing integer division with the same variables and printing its result
