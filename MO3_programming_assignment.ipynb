9.2

```python
def get_odds(first=0, last=10, step=1):
    number = first
    while number < last:
        if number % 2 == 1:
            yield number
        number += step
        
my_odds = get_odds()
count = 0

for odds in my_odds:
    count += 1
    if count >= 3:
        print(f"The third value is: {odds}")
        break
    
print("program has finished")
```

The third value from the odd numbers has been printed
